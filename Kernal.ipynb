```
import pandas as pd 
import io
from google.colab import files

uploaded = files.upload()
```

```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
data = io.BytesIO(uploaded['Bankwise Gross NPL Ratio.xlsx'])
```

```
df = pd.read_excel(data , sheet_name = 'Sheet2',  skiprows = 0)
# displaying first few rows of the dataset
df.head()
```

```
# List of dates to plot
dates = ['30.09.17','31.12.17','30.09.18','31.12.18','30.09.19']

# Iterate through the dates
for date in dates:
    # Subset to the date
    subset = df[df['Date'] == date]

   #kernel : {‘gau’ | ‘cos’ | ‘biw’ | ‘epa’ | ‘tri’ | ‘triw’ } 

    # Draw the density plot
    sns.kdeplot(subset['NPL'],  kernel = 'gau', 
                 gridsize=100, cumulative= False, legend = True,
                 Label = date)
    
# Plot formatting
plt.legend(prop={'size': 16}, title = 'Date')
plt.title('TITLE')
plt.xlabel('X')
plt.ylabel('Density')
plt.ylim(-.05, 0.25)
plt.xlim(0, 40)
```
